   ## Web-scraping de dados metereológicos

```python
# Importando as bibliotecas necessárias 
import time
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import ast
import pandas as pd
```

### Obtendo os dados de previsão metereológica

Dados retirados do site climatempo para a cidade de Araranguá - SC 

```python
#abrindo o navegador no site solicitado
navegador = webdriver.Chrome()
site  = 'https://www.climatempo.com.br/previsao-do-tempo/amanha/cidade/2067/ararangua-sc'
navegador.get(site)
navegador.implicitly_wait(20)


#requisição html do site
html_content = requests.get(site).text

#retirando os dados do html 
soup = BeautifulSoup(html_content, "lxml")

dados=[]
for i in soup.find_all("div", attrs={"class": "wrapper-chart"}):
    dados.append(i['data-infos'])

# processando a sintaxe de gramatica 
dados = ast.literal_eval(dados[0])

#Imprimindo a coluna dos dados retirados da página
dados[0].keys()
```

```python
# tratando os dados para uma lista
dados_previsao=[]
for ind in range(len(dados)):
    dicio={}
    for k,v in dados[ind].items():
        if k in "date":
            dicio[k]=v
        elif k in ["wind","temperature","rain"]:
            for k,v in dados[ind][k].items():
                dicio[k]=v
    dados_previsao.append(dicio)

# Imprimindo os dados de previsão 
dados_previsao


```

```python
# Importando os dados para uma estrutura de dataframe 
previsao_df = pd.DataFrame(dados_previsao)
# Imprimindo os dados 
previsao_df
    
```

```python
#Imprimindo o formato dos dados
print(previsao_df.dtypes)
```

```python
#convertendo os dados de tempo para datetime
previsao_df['date'] = pd.to_datetime(previsao_df['date'])
previsao_df['horas']= [d.time() for d in previsao_df['date']] #criando coluna de horas

previsao_df.set_index('horas',inplace=True) #selecionando novo índice

cols = previsao_df.columns.tolist() #alterando a ordem das colunas
cols = cols[-2:] + cols[1:5] + cols[:-6]
previsao_df = previsao_df[cols]

#convertendo os dados do tipo string para o tipo float
previsao = previsao_df[previsao_df.columns[:-2]].astype(float)
previsao['direction']=previsao_df['direction']
previsao['date'] = previsao_df['date']

#imprimindo o dataframe de dados de previsão 
previsao
```

### Previsões 

```python

print(f"A precipitação acumulada para o dia {previsao['date'][0].day}/{previsao['date'][0].month} é de: {previsao['precipitation'].sum()} mm ")

if previsao['precipitation'].max() != 0 :
    print("A precipitação máxima é de {previsao['precipitation'].max()} as {previsao['precipitation'].idxmax()}")
    
print(f"A temperatura mínima prevista é de: {previsao['temperature'].min()} °C")

print(f"A temperatura máxima prevista é de: {previsao['temperature'].max()} °C")

idmaxw = previsao['gust'].idxmax()                                                                                                                      
                                                                                                            
print(f"A velocidade máxima do vento é de {previsao['velocity'].max()} km/h e rajadas de {previsao['gust'].max()} km/h de direção {previsao['direction'][idmaxw]} a {previsao['direction_degrees'][idmaxw]}°")
```

### Plotando os dados obtidos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
%matplotlib inline
```

```python
previsao['h']=previsao['date'].apply(lambda time: time.hour)
previsao
```

```python
#Plotando a previsão de temperatura
sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(1, 1,figsize=(12,6))
ax.plot(previsao['h'], previsao['temperature'],lw=2,marker='.', markersize=15,color='r',)
xticks=range(0,24,1)
ax.set_xticks(xticks)
ax.set_title("Previsão de temperatura", fontsize=16)
plt.xlabel("horário",fontsize=14)
plt.ylabel("temperatura °C",fontsize=14)
ax.set_xlim(-0.2,23.5)
```

```python
#Plotando a previsão de vento
fig, ax = plt.subplots(1, 1,figsize=(12,6))
previsao['gust'].plot.area(alpha=0.6)
previsao['velocity'].plot.area(alpha=0.7)
ax.set_xlim(0,24)
ax.set_xticks(range(0,86400,7200))
ax.set_title("Previsão de vento", fontsize=18)
plt.xlabel("horário",fontsize=14)
plt.ylabel("Velocidade km/h",fontsize=14)
plt.legend(["Rajadas","Velocidade"],bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
```

```python
#Verificando se há ou não previsão de chuva para realizar o plot
if previsao['precipitation'].sum() != 0:
    previsao['precipitation'].plot(kind='bar')
else:
    fig, ax = plt.subplots(1, 1,figsize=(8,6))
ax.annotate('Sem previsão de chuva',
            xy=(.3, .575), xycoords='figure fraction',
            horizontalalignment='left', verticalalignment='top',
            fontsize=20)
    
```
